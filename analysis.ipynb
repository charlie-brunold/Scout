```python
# function that creates a team-specific dataframe, given the team
def strip_suffix(x, suff):
    # remove the suffix from the column name
    if x.endswith(suff):
        return x[:-len(suff)]
    else:
        return x
    
visit_cols = [col for col in df.columns if not col.endswith('_h')] # all the visiting team columns
visit_cols_no_suff = [strip_suffix(col, '_v') for col in visit_cols] # remove the _v suffix
home_cols = [col for col in df.columns if not col.endswith('_v')] # all the home team columns
home_cols_no_suff = [strip_suffix(col, '_h') for col in home_cols] # remove the _h suffix

def create_team_df(team):
    df_team_v = df[(df.team_v == team)]
    opponent = df_team_v['team_h']
    df_team_v = df_team_v[visit_cols]
    df_team_v.columns = visit_cols_no_suff
    df_team_v['home_game'] = 0
    df_team_v['opponent'] = opponent

    df_team_h = df[(df.team_h == team)]
    opponent = df_team_h['team_v']
    df_team_h = df_team_h[home_cols]
    df_team_h.columns = home_cols_no_suff
    df_team_h['home_game'] = 1
    df_team_h['opponent'] = opponent

    df_team = pd.concat((df_team_v, df_team_h))
    df_team.sort_values(['date', 'game_no'], inplace=True)

    for winsize in [162, 30]: # past 162 games, past 30 games
        suff = str(winsize) # suffix to add to the column names (rolling sum for 162 games and 30 games)
        for raw_col in ['AB', 'H', '2B', '3B', 'HR', 'BB', 'runs', 'SB', 'CS', 'ERR']:
            new_col = 'rollsum_' + raw_col + '_' + suff
            df_team[new_col] = df_team[raw_col].rolling(window=winsize, closed='left').sum()

        df_team['rollsum_AVG_' + suff] = df_team['rollsum_H_' + suff] / df_team['rollsum_AB_' + suff]
        df_team['rollsum_OBP_' + suff] = (df_team['rollsum_H_' + suff] + df_team['rollsum_BB_' + suff]) / (df_team['rollsum_AB_' + suff] + df_team['rollsum_BB_' + suff])
        df_team['rollsum_SLG_' + suff] = ((df_team['rollsum_H_' + suff] + 2 * df_team['rollsum_2B_' + suff] + 3 * df_team['rollsum_3B_' + suff] + 4 * df_team['rollsum_HR_' + suff])) / df_team['rollsum_AB_' + suff]
        df_team['rollsum_OPS_' + suff] = df_team['rollsum_OBP_' + suff] + df_team['rollsum_SLG_' + suff]
    
    df_team['season_game'] = df_team['season']*1000 + df_team['game_no']
    df_team.set_index('season_game', inplace=True)
    return (df_team)
```

```python
# create a dictionary of the teams mapped to their data frames
team_data_dict = {}
# for only teams in BOTH team_v unique and team_h unique
for team in df.team_v.unique():
    if team in df.team_h.unique() and team != 'SPU':
        # print(team)
        # create the team dataframe
        team_data_dict[team] = create_team_df(team)
```

```python
AVG_162_h = np.zeros(df.shape[0]) # feature vectors
AVG_162_v = np.zeros(df.shape[0])
OBP_162_h = np.zeros(df.shape[0])
OBP_162_v = np.zeros(df.shape[0])
SLG_162_h = np.zeros(df.shape[0])
SLG_162_v = np.zeros(df.shape[0])
OPS_162_h = np.zeros(df.shape[0])
OPS_162_v = np.zeros(df.shape[0])
SB_162_h = np.zeros(df.shape[0])
SB_162_v = np.zeros(df.shape[0])
CS_162_h = np.zeros(df.shape[0])
CS_162_v = np.zeros(df.shape[0])
ERR_162_h = np.zeros(df.shape[0])
ERR_162_v = np.zeros(df.shape[0])
AVG_30_h = np.zeros(df.shape[0])
AVG_30_v = np.zeros(df.shape[0])
OBP_30_h = np.zeros(df.shape[0])
OBP_30_v = np.zeros(df.shape[0])
SLG_30_h = np.zeros(df.shape[0])
SLG_30_v = np.zeros(df.shape[0])
OPS_30_h = np.zeros(df.shape[0])
OPS_30_v = np.zeros(df.shape[0])
SB_30_h = np.zeros(df.shape[0])
SB_30_v = np.zeros(df.shape[0])
CS_30_h = np.zeros(df.shape[0])
CS_30_v = np.zeros(df.shape[0])
ERR_30_h = np.zeros(df.shape[0])
ERR_30_v = np.zeros(df.shape[0])

# loop through the data frame and fill in the feature vectors
i = 0
for index, row in df.iterrows():
    if i%1000 == 0:
        print(i)
    home_team = row['team_h']
    away_team = row['team_v']
    game_index_v = row['season']*1000 + row['game_no_v']
    game_index_h = row['season']*1000 + row['game_no_h']
    AVG_162_h[i] = team_data_dict[home_team].loc[game_index_h, 'rollsum_AVG_162']
    AVG_162_v[i] = team_data_dict[away_team].loc[game_index_v, 'rollsum_AVG_162']
    OBP_162_h[i] = team_data_dict[home_team].loc[game_index_h, 'rollsum_OBP_162']
    OBP_162_v[i] = team_data_dict[away_team].loc[game_index_v, 'rollsum_OBP_162']
    SLG_162_h[i] = team_data_dict[home_team].loc[game_index_h, 'rollsum_SLG_162']
    SLG_162_v[i] = team_data_dict[away_team].loc[game_index_v, 'rollsum_SLG_162']
    OPS_162_h[i] = team_data_dict[home_team].loc[game_index_h, 'rollsum_OPS_162']
    OPS_162_v[i] = team_data_dict[away_team].loc[game_index_v, 'rollsum_OPS_162']
    SB_162_h[i] = team_data_dict[home_team].loc[game_index_h, 'rollsum_SB_162']
    SB_162_v[i] = team_data_dict[away_team].loc[game_index_v, 'rollsum_SB_162']
    CS_162_h[i] = team_data_dict[home_team].loc[game_index_h, 'rollsum_CS_162']
    CS_162_v[i] = team_data_dict[away_team].loc[game_index_v, 'rollsum_CS_162']
    ERR_162_h[i] = team_data_dict[home_team].loc[game_index_h, 'rollsum_ERR_162']
    ERR_162_v[i] = team_data_dict[away_team].loc[game_index_v, 'rollsum_ERR_162']
    AVG_30_h[i] = team_data_dict[home_team].loc[game_index_h, 'rollsum_AVG_30']
    AVG_30_v[i] = team_data_dict[away_team].loc[game_index_v, 'rollsum_AVG_30']
    OBP_30_h[i] = team_data_dict[home_team].loc[game_index_h, 'rollsum_OBP_30']
    OBP_30_v[i] = team_data_dict[away_team].loc[game_index_v, 'rollsum_OBP_30']
    SLG_30_h[i] = team_data_dict[home_team].loc[game_index_h, 'rollsum_SLG_30']
    SLG_30_v[i] = team_data_dict[away_team].loc[game_index_v, 'rollsum_SLG_30']
    OPS_30_h[i] = team_data_dict[home_team].loc[game_index_h, 'rollsum_OPS_30']
    OPS_30_v[i] = team_data_dict[away_team].loc[game_index_v, 'rollsum_OPS_30']
    SB_30_h[i] = team_data_dict[home_team].loc[game_index_h, 'rollsum_SB_30']
    SB_30_v[i] = team_data_dict[away_team].loc[game_index_v, 'rollsum_SB_30']
    CS_30_h[i] = team_data_dict[home_team].loc[game_index_h, 'rollsum_CS_30']
    CS_30_v[i] = team_data_dict[away_team].loc[game_index_v, 'rollsum_CS_30']
    ERR_30_h[i] = team_data_dict[home_team].loc[game_index_h, 'rollsum_ERR_30']
    ERR_30_v[i] = team_data_dict[away_team].loc[game_index_v, 'rollsum_ERR_30']
    i += 1


```
